```python
def scrape_month_temp(month = None):

    import requests
    import pandas as pd
    from bs4 import BeautifulSoup

    if month == None:
        my_url = "https://www.extremeweatherwatch.com/us-state-averages"
    else: 
        my_url = "https://www.extremeweatherwatch.com/us-state-averages" + "/month-" + month

    response = requests.get(my_url)
    soup = BeautifulSoup(response.content, "html.parser")

    table = soup.find('table', class_='bordered-table sort-table')

    headers = [header.text for header in table.find_all('th')]

    # Extract all rows of the table
    rows = table.find_all('tr')[1:]  # Skip header row
    
    # Initialize list to store data
    data = []

    for row in rows:

        cells = row.find_all('td')
        state_tag = cells[0]
        state = state_tag.get_text(strip = True)
        high_tag = cells[1]
        high = high_tag.get_text(strip = True)
        low_tag = cells[2]
        low = low_tag.get_text(strip = True)
        precip_tag = cells[3]
        precip = precip_tag.get_text(strip = True)

        data.append({
                headers[0]: state,
                headers[1]: high,
                headers[2]: low,
                headers[3]: precip
    })
    df = pd.DataFrame(data)
    return df
```

```python
scrape_month_temp()
```

```python
import pandas as pd
summer = ["may","june", "july", "august"]
summerdf = pd.DataFrame(summer)
summerdf.apply(scrape_month_temp())
```
