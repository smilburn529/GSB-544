```python
def sing_verse(num):
  song = str(num) + " bottles of beer on the wall \n" + str(num) + " bottles of beer \n" + " take one down, pass it around, \n" + str(num-1) + " bottles of beer on the wall \n"
  
  return song
```

```python
song = list()

for i in range(100,97,-1):
  song.append(sing_verse(i))
  
print(song)
```
