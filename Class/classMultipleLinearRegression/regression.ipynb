```python
from palmerpenguins import load_penguins
import pandas as pd

# Load the dataset
penguins = load_penguins()

# Check the unique values in the species column
print(penguins['species'].unique())
```

```python
# Using get_dummies on the 'species' column
species_dummies = pd.get_dummies(penguins['species'], prefix='species')
print(species_dummies.head())
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder()
enc.fit(penguins[["species"]])

enc.transform(penguins[["species"]]).toarray()

```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit_transform(penguins[["bill_length_mm"]])
```

```python
# Using OneHotEncoder with the correct parameter
encoder = OneHotEncoder(sparse_output=False)  # Use sparse_output instead of sparse
species_encoded = encoder.fit_transform(penguins[['species']])

# Convert the encoded array to a DataFrame for easy comparison
species_encoded_df = pd.DataFrame(species_encoded, columns=encoder.get_feature_names_out(['species']))
species_encoded_df.head()
```
